Question : Write a python to demonstrate polymorphism where two different classes have the same method name,and common functions cells those method.

```python
class Cat:
  def sound(self):
    return "Meow"

class Dog:
  def sound(self):
    return "Woof"

c=Cat()
d=Dog()

# using polymorphism
animals=[c,d]
for animal in animals:
  print(animal.sound())
```

```python
class Cow:
  def sound(self):
    return "Moo"

class Sheep:
  def sound(self):
    return "Baa"

c=Cow()
s=Sheep()

#using polymorphism

animals=[c,s]
for animal in animals:
  print(animal.sound())
```

```python
class Vehicle:
  def speed(self):
    return "The vehicle move at standard speed."

class Car(Vehicle):
  def speed(self):
    return "The car move at a speed 100 km/h."


v=Vehicle()
c=Car()

print(c.speed())
print(v.speed())
```
